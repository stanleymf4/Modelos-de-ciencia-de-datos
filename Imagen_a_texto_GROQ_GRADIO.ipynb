En esta iniciativa se realiz√≥ la ejecuci√≥n de un modelo de Groq para la descripci√≥n autom√°tica de im√°genes mediante texto. Para facilitar la interacci√≥n con el usuario se integr√≥ Gradio, creando una interfaz sencilla en Google Colab que permite cargar una imagen, procesarla con el modelo y obtener una descripci√≥n de su contenido.

Este tipo de soluciones, aunque simples en su concepci√≥n, tienen un amplio potencial de aplicaci√≥n en sectores como la agricultura, la prevenci√≥n de riesgos o la gesti√≥n de desastres. Sin embargo, para alcanzar mayor precisi√≥n en escenarios espec√≠ficos, es recomendable aplicar t√©cnicas de Fine-Tuning, entrenando el modelo con im√°genes etiquetadas que respondan a necesidades particulares.

üîß Tecnolog√≠as utilizadas:

**Google Colab** ‚Üí entorno de laboratorio para pruebas.

**Gradio** ‚Üí creaci√≥n de la interfaz de usuario para carga y salida de im√°genes.

**Groq** ‚Üí plataforma de inferencia en la que est√° publicado el modelo de IA utilizado.

**ChatGPT** ‚Üí apoyo como asistente en la generaci√≥n y ajuste de c√≥digo.

üí° Nota: para utilizar Groq es necesario registrarse en la plataforma y generar una API Key, la cual debe configurarse de manera segura en los secrets de Google Colab.


```python
# 1. Instalaci√≥n de dependencias
!pip install groq -q

from groq import Groq
import base64
import os
from google.colab import userdata

# Se debe tener configurado un api key en Groq y configurarla como secreto en google colab
client = Groq(api_key=userdata.get('GROQ_KEY'))

# Esta funci√≥n permite c√≥dificar la imagen en base 64
def encode_image(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')

# Esta funci√≥n tiene la l√≥gica para interactuar con el modelo y mostrar la descripci√≥n de la imagen
def description_image(image_path):

  # obtener la igagen codificada
  base64_image = encode_image(image_path)

  # llamada al modelo e instrucci√≥n para interactuar con el modelo con una salida tipo chat
  chat_completion = client.chat.completions.create(
      messages=[
          {
              "role": "user",
              "content": [
                  {"type": "text", "text": "What's in this image?"},
                  {
                      "type": "image_url",
                      "image_url": {
                          "url": f"data:image/jpeg;base64,{base64_image}",
                      },
                  },
              ],
          }
      ],
      model="meta-llama/llama-4-scout-17b-16e-instruct",
  )

  return chat_completion.choices[0].message.content
```

```python
import gradio as gr
from PIL import Image

# Funci√≥n para recibir y mostrar la imagen
def procesar_imagen(img):
    img_evaluada = description_image(img)
    return img_evaluada

# Crear la interfaz
iface = gr.Interface(
    fn=procesar_imagen,        # funci√≥n que procesa la imagen
    inputs=gr.Image(type="filepath"),
    outputs=gr.Textbox(label="Texto", lines=10, placeholder="Aqu√≠ aparecer√° la descripci√≥n..."),
    title="Cargar y mostrar imagen",
    description="Sube una imagen desde tu PC para visualizarla en Colab."
)

# Lanzar la interfaz
iface.launch(debug=True)

```
